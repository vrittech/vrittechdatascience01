```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the data for the box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
print(data)
```

```python
len(data)
```

```python
print(data[0].max())
print(data[0].min())
print(data[0].mean())
```

```python
print(data[1].max())
print(data[1].min())
print(data[1].mean())
```

```python
print(data[2].max())
print(data[2].min())
print(data[2].mean())
print(np.median(data[2]))
```

```python
fig, ax = plt.subplots()

# Generate the box plot
ax.boxplot(data)

# Customize the chart
ax.set_title('Box Plot', fontsize=16, fontweight='bold')
ax.set_xticklabels(['Data 1', 'Data 2', 'Data 3'])
ax.set_xlabel('Data')
ax.set_ylabel('Value')

plt.show()
```
