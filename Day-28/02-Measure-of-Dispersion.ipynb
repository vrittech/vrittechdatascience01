```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color = 'blue')
```

## Variance

```python
variance = np.var(x)
print("Variance:", variance)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = x.mean(), color='yellow')
plt.axvline(x = variance, color='red')
```

## Standard Deviation

```python
std = np.std(x)
print("Standard Deviation:", std)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = std, color='red')
plt.axvline(x = x.mean(), color='yellow')
```
